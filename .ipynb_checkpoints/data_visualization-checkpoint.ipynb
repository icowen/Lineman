```python
import re
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import HTML
from matplotlib.animation import FuncAnimation
```

```python
data = pd.read_csv('netdata.csv')
data.head()
```

```python
data.drop('X', inplace=True, axis=1)
data.fillna('Run', inplace=True)
print(f'New dimensions of DataFrame: {data.shape}')
```

```python
num_plays = data["playId"].unique().size
print(f'Number of unique plays: {num_plays}')
print(f'Average number of frames per play {round(data.shape[0] / num_plays, 2)}')
```

```python
gameId = 2017090700
playId = 68
play = data.loc[(data["playId"] == playId) & (data["gameId"] == gameId)]
```

```python
# Create figure
fig, ax = plt.subplots()
fig.set_tight_layout(True)
plt.xticks([], [])
plt.yticks([], [])

# Draw field dimensions
field_length = 120
field_width = 160 / 3
ax.set_xlim([0, field_length])
ax.set_ylim([0, field_width])
ax.set_facecolor('green')

# Add end zones
ax.fill_between([0, 10], field_width, color='maroon')
ax.fill_between([110, field_length], field_width, color='gold')

# Add yard lines and markings
for i in range(1, 12):
    yard_line = i * 10
    next_yard_line = (i + 1) * 10 + 1
    ax.axvline(yard_line, c='white', zorder=0)
    if i < 6: ax.annotate(yard_line, (next_yard_line, 5), c='white')
    elif i < 10: ax.annotate(100 - yard_line, (next_yard_line, 5), c='white')

# Overlay with an empty scatter plot
scat = ax.scatter([], [], s=16, edgecolors='black')
plt.show()
```

```python
def get_new_x_y(frame_df):
    # Find columns with coordinate data
    x_cols = [c for c in frame_df.columns if re.match('.*_x$|^X_', c)]
    y_cols = [c for c in frame_df.columns if re.match('.*_y$|^Y_', c)]
    
    # Get coordinates for each player in the frame
    x = [frame_df[x].values[0] for x in x_cols]
    y = [frame_df[y].values[0] for y in y_cols]
    
    # Identify a team for each pair of coordinates
    labels = ['OL'] * 5 + ['def'] * 5 + ['off'] * 6 + ['def'] * 11
    
    # Return coordinate/team data frame
    return pd.DataFrame(dict(x=x, y=y, label=labels))


def update(frame):
    # Get data for this frame 
    frame_df = play[play["frame.id"] == frame]
    
    # Clean coordinates
    df = get_new_x_y(frame_df)
    
    # Convert coordinate data frame to list of tuples and plot
    x_y = [(x, y) for x, y in zip(df['x'], df['y'])]
    scat.set_offsets(x_y)
    scat.set_edgecolors('black')
    
    # Identify offense vs defense
    colors = ['r' if x == 'def' else 'b' for x in df['label']]
    scat.set_color(colors)
    
    # Update frame label
    ax.set_xlabel(f'Frame {frame}')


anim = FuncAnimation(fig, update, frames=play["frame.id"], interval=100)
```

```python
HTML(anim.to_jshtml())
```
